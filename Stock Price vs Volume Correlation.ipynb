```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Read stock data from a CSV file
stock_data = pd.read_csv('WMT.csv')

# Extract 'Volume' and 'Close' columns
volume = stock_data['Volume'].values.reshape(-1, 1)  # Reshape to a 2D array
close_price = stock_data['Close'].values

# Calculate correlation coefficient
correlation_coefficient = stock_data['Volume'].corr(stock_data['Close'])

# Plot volume against price
plt.figure(figsize=(10, 6))
plt.scatter(volume, close_price, color='blue', alpha=0.5)
plt.title('Stock Volume vs Price for Walmart')
plt.xlabel('Volume')
plt.ylabel('Price ($)')

# Perform linear regression
regression_model = LinearRegression()
regression_model.fit(volume, close_price)

# Predict prices using the regression model
predicted_prices = regression_model.predict(volume)

# Plot the regression line
plt.plot(volume, predicted_prices, color='red')

# Print regression coefficients
print("Intercept:", regression_model.intercept_)
print("Coefficient:", regression_model.coef_[0])

# Print correlation coefficient
print("Correlation Coefficient:", correlation_coefficient)

plt.show()
```
