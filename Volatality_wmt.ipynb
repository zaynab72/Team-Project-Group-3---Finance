```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
file_path = '../data/WMT.csv'  
df = pd.read_csv(file_path)

# Calculate the average of 'HIgh' and 'Low' for each row
df['Average_Stock_Price'] = (df['High'] + df['Low']) / 2

print("Average stock price for each row:")
print(df['Average_Stock_Price'])
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Average_Stock_Price'], color='blue', label='Average Stock Price')
plt.xlabel('Date')
plt.ylabel('Average Stock Price')
plt.title('Average Stock Price Over Time')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot volume against date
plt.figure(figsize=(10, 6))
plt.bar(df['Date'], df['Volume'], color='green', label='Volume')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.title('Volume Over Time')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Average_Stock_Price'], color='blue', label='Average Stock Price')
plt.xlabel('Date')
plt.ylabel('Average Stock Price')
plt.title('Average Stock Price and Volume Over Time')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)

# Create the second plot for volume and overlay it on the first plot
plt.twinx()
plt.bar(df['Date'], df['Volume'], color='green', alpha=0.5, label='Volume')
plt.ylabel('Volume')
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
import numpy as np
# Calculate Returns and Volatility
df['Return'] = df['Close'].pct_change()
df['Volatility'] = df['Return'].rolling(window=20).std() * np.sqrt(20)

# Plot Volatility
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Volatility'], label='Volatility', color='red')
plt.title('Volatility Over Time (20-day Rolling Standard Deviation)')
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.legend()
plt.show()
```
