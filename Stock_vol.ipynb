```python
# Importing of Python Pandas Module
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime
```

```python
# Creating Data frame for each of the 4 selcted stocks from the CSV files
# Note: the stock ticker symbol preceeds the df notation eg: wmf_df is the Wallmart Data Frame
aci_df = pd.read_csv("Data/ACI.csv")
cost_df = pd.read_csv("Data/COST.csv")
tgt_df = pd.read_csv("Data/TGT.csv")
wmt_df = pd.read_csv("Data/WMT.csv")
```

```python
# Function to plot Stock Volume over 5 Years
def plot_stock_vol(symbol, ticker_df):
    
    #  Converting y axis units to Millions for easier interpretation
    y_axis = ticker_df['Volume'] / 1000000
    
    # Changing Object field to Date and then removing the day to display month and year in x axis
    ticker_df['Date'] = pd.to_datetime(ticker_df['Date'])
    x_axis = ticker_df['Date'].dt.strftime('%b/%Y')

    # Plotting of bar chart
    plt.bar( x_axis, y_axis, width=0.7)
    
    
    # Formatting of bar chart
    plt.title(f"{symbol} Stock Volume Change over 5 year")
    plt.xlabel("Time (Months)")
    plt.ylabel("Stock Volume Change (Millions)")
    plt.xticks(rotation=90)
    plt.xticks(fontsize=6)
    plt.tight_layout()

    # Saving of chart to output folder
    plt.savefig(f"../output/{symbol} Stock Vol Plot.png")
```

```python
# Plotting of COST Stock Vol over 5 years
plot_stock_vol(symbol = "COST",
               ticker_df = cost_df)
```

```python
# Plotting of ACI Stock Vol over 5 years
plot_stock_vol(symbol = "ACI",
               ticker_df = aci_df)
```

```python
# Plotting of TGT Stock Vol over 5 years
plot_stock_vol(symbol = "TGT",
               ticker_df = tgt_df)
```

```python
# Plotting of WMT Stock Vol over 5 years
plot_stock_vol(symbol = "WMT",
               ticker_df = wmt_df)
```
