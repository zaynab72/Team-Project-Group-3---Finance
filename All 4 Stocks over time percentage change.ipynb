```python
# Create libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('../data/COST.csv')

# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Set the 'Date' column as the index
df.set_index('Date', inplace=True)

# Resample the data to monthly frequency and calculate the mean
monthly_df = df.resample('M').mean()

# Reset index to make 'Date' a column again
monthly_df.reset_index(inplace=True)

# Display the "raw" values
display(monthly_df)
```

```python
# Calculate Percentage Change in Stock Price

# Read the daily stock data from the CSV file
stock_data = pd.read_csv('../data/COST.csv')

# Convert the date column to datetime format
stock_data['Date'] = pd.to_datetime(stock_data['Date'])

# Group the data by month
monthly_data1 = stock_data.groupby(pd.Grouper(key='Date', freq='M')).agg({
    'Open': 'first',
    'Close': 'last'
}).reset_index()

# Calculate the price difference between closing and opening prices
monthly_data1['Price_Difference'] = monthly_data1['Close'] - monthly_data1['Open']

# Calculate the percentage change
monthly_data1['Percentage_Change'] = (monthly_data1['Price_Difference'] / monthly_data1['Open']) * 100

# Display the percentage change
display(monthly_data1)
```

```python
import pandas as pd

# Read the daily stock data from the CSV files for each stock
stock_data_cost = pd.read_csv('../data/COST.csv')
stock_data_wmt = pd.read_csv('../data/WMT.csv')
stock_data_tgt = pd.read_csv('../data/TGT.csv')
stock_data_aci = pd.read_csv('../data/ACI.csv')

# Add a 'Symbol' column to each DataFrame
stock_data_cost['Symbol'] = 'COST'
stock_data_wmt['Symbol'] = 'WMT'
stock_data_tgt['Symbol'] = 'TGT'
stock_data_aci['Symbol'] = 'ACI'

# Concatenate the individual DataFrames into one
combined_data = pd.concat([stock_data_cost, stock_data_wmt, stock_data_tgt, stock_data_aci])

# Convert the date column to datetime format
combined_data['Date'] = pd.to_datetime(combined_data['Date'])

# Group the data by month and calculate percentage change for each stock
monthly_data = combined_data.groupby(['Symbol', pd.Grouper(key='Date', freq='M')]).agg({
    'Open': 'first',
    'Close': 'last'
}).reset_index()
monthly_data['Price_Difference'] = monthly_data['Close'] - monthly_data['Open']
monthly_data['Percentage_Change'] = (monthly_data['Price_Difference'] / monthly_data['Open']) * 100

# Save the combined data to a CSV file
monthly_data.to_csv('../Data/combined_data_with_symbol.csv', index=False)

# Display the first few rows of the combined data
print(monthly_data.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the combined data from the CSV file
monthly_data = pd.read_csv('../Data/combined_data_with_symbol.csv')

# Convert the 'Date' column to datetime format
monthly_data['Date'] = pd.to_datetime(monthly_data['Date'])

# Plot the percentage change in stock prices over time for each company
plt.figure(figsize=(10, 6))

for symbol in monthly_data['Symbol'].unique():
    company_data = monthly_data[monthly_data['Symbol'] == symbol]
    plt.plot(company_data['Date'], company_data['Percentage_Change'], label=symbol)

plt.title('Monthly Percentage Change in Stock Prices')
plt.xlabel('Date')
plt.ylabel('Percentage Change')
plt.legend()
plt.grid(True)


# Saving of chart to output folder
plt.savefig("../output/All Stock Percent Change - Line Chart.png")

plt.show()
```
